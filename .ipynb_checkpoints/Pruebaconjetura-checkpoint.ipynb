```python
from itertools import product
import numpy as np
import itertools
from itertools import combinations
from copy import deepcopy
from operator import add,mul
from collections import defaultdict
from collections import Counter
from sage.combinat.cartesian_product import CartesianProduct_iters
import time
#la siguiente funcion recibe una lista y un elemento de ella, y nos da de izquierda a derecha
#los anteriores elemento de la lista mas cercanos, uno mayor y otro menor, 
#si no hay el menor es 0 o el mayor es el tamaño de la lista 

#este programa servirá para obtener la subpalabra de sigma donde necesitamos hacer la división
#dandonos los indices del primer elemento de esta subpalabra y el último

def intervalo(numbers, target):
    smaller = None
    larger = None

    for num in numbers:
        
        if num == target:
            break
        if num < target:
            if smaller is None or num >=  smaller:
                smaller = num+1
        elif num > target:
            if larger is None or num  < larger:
                larger = num+1
    
    if smaller==None:
        smaller=0
    if larger ==None:
        larger=len(numbers)+1
    return [smaller, larger]

#Una vez obtenidos la palabra a dividir utilizando los indices dado en el anterior programa 
#hacemos recursivamente las debidas divisiones

def conjuntos(sigma,tau):
    particiones_recursivas=[]
    for t in tau:
        I=intervalo(tau,t)
        particion_t=[sigma[I[0]:t+1],sigma[t+1:I[1]]]
        particiones_recursivas.append(particion_t)
    return particiones_recursivas



#clasifica los arboles binario usando la cantidad caracteres en cada división en cada nodo, 
#Alinea todos los arboles a la derecha
#Despues caclcula para cada una de las clasificaiones antreriores su clase de syvlester
def clases_normales(n):
    estruct_normals = set()
    for arbol in BinaryTrees(n).list():
        estruct_normals.add(OrderedTree(arbol).normalize())

    clases_sylves = []

    
    for arbolnormal in estruct_normals:
        clase_sylv = [list(reversed(tb)) for tb in BinaryTree(arbolnormal).sylvester_class()]
        clases_sylves.append(clase_sylv)

    return clases_sylves


#Recibe un elemento A de una de las listas de clases_normales(n) crea unas permutaciones de S_n+1 
#dads por los elementos de conjuntos([0,...,n],A) que tienen la misma cantidad de elementos.
#se vuelven a obtener elementos de la misma clase
def permu_equiva(estructura):
    A=list(range(1, len(estructura)+2))
    B=[]
    for i in conjuntos(A,estructura):
        if len(i[0])==len(i[1]):
            B.append(list(zip(i[0],i[1])))
    return B

#crea los cosets dados por el subgrupo generado por los elementos obtenidos por permu_equiva (lo cual recibe la funcion)

def clases_sigma(permudeequiv,n):
    
    G =  SymmetricGroup(n+1)

    quarter_turn = PermutationGroup(permudeequiv)
    S = G.subgroup(quarter_turn)
    rc = G.cosets(S)
    clases=[]
    for clase_equiv in rc:
        first_vert= list(range(1, n+2))
        result = [clase_equiv[0](i)-1 for i in first_vert]
        clases.append(result)
    return clases





def representantes_de_clase(n):
    normal_tree=clases_normales(n)
    sumvol=[]
    for i in normal_tree:
        tau=[k-1 for k in i[0]]
        TT= clases_sigma(permu_equiva(tau),n)
        
        for m in TT:
            for j in i:
                l=[k-1 for k in j]
                sumvol.append([ m,l])
        
        
    return sumvol


def vector_principal(i,j,n):
    w = zero_vector(SR, n+1)
    w[min(i,j)]=var('p{}{}'.format(min(i,j),max(i,j)))
    w[max(i,j)]=-var('p{}{}'.format(min(i,j),max(i,j)))

    return w


def gram(n):
    P=matrix(SR,n)
    for i in range(1,n+1):
        for j in range(1,n+1):
            if i==j:
                s=0                    
                for l in range(n+1):
                        if l!=i:
                            s=s+var('p{}{}'.format(min(i,l),max(i,l))) 
                P[i-1,j-1]=s
            else:
                P[i-1,j-1]=-var('p{}{}'.format(min(i,j),max(i,j)))  
                P[j-1,i-1]=P[i-1,j-1]
    return P

def volumen(sigma,tau):
    volum=1
    for i in conjuntos(sigma,tau):
        C = [[x, y] for x in i[0] for y in i[1]]
        factor_volumen = sum([var('p{}{}'.format(min(c[0],c[1]),max(c[0],c[1]))) for c in C])
        volum *= factor_volumen
    return volum.expand()

def volumen_prueba(sigma,tau):
    volum=1
    a=conjuntos(sigma,tau)
    b=[[x, y] for x in a[0][0] for y in a[0][1]]
    for i in a:
        C = [[x, y] for x in i[0] for y in i[1]]
        factor_volumen = sum([var('p{}{}'.format(min(c[0],c[1]),max(c[0],c[1]))) for c in C])
        volum *= factor_volumen
    return (volum*sum([var('p{}{}'.format(min(c[0],c[1]),max(c[0],c[1]))) for c in b])).expand()


def volumentotal(n):
    suma_total=0
    A=representantes_de_clase(n)
    for i in A:
        suma_total+=volumen(i[0],i[1])
    return suma_total.expand()






def DERIVADAS1(dimension,D):
    Derivadas=matrix(SR,dimension+1)
    
    for i in range(dimension+1):
        for j in range(dimension+1):
            if i==j:
                Derivadas[i,j]=0
            else:
                Derivadas[i,j]=D.diff(var('p{}{}'.format(min(i,j),max(i,j))) )
    return(Derivadas)

def DERIVADAS2(dimension,D):
    Derivadas=matrix(SR,dimension+1)
    
    for i in range(dimension+1):
        for j in range(dimension+1):
            if i==j:
                Derivadas[i,j]=0
            else:
                Derivadas[i,j]=D.diff(var('D{}{}'.format(min(i,j),max(i,j))) )
    return(Derivadas)

def DERIVADAS(dimension):
    Derivadas=matrix(SR,dimension+1)
    
    for i in range(dimension+1):
        for j in range(dimension+1):
            if i==j:
                Derivadas[i,j]=0
            else:
                Derivadas[i,j]=var('D{}{}'.format(min(i,j),max(i,j))) 
    return(Derivadas)



#La norma de un vector en y-coordenadas

def norma1(n,vector1):
    N1=0
    dif=DERIVADAS(n)
    for i in range(1,n+1):
        N1=N1+(dif[0][i])*(vector1[i]*vector1[i]).expand()
    N2=0
    for i in range(1,n+1):
        for j in range(1,n+1):
            if i<j:
                N2=N2+((dif[0][i])+(dif[0][j])
                       -(dif[i][j]))*(vector1[i]*vector1[j]).expand()

    norma=(N1+N2)
    return norma


def e_conjuntos(sigma,tau):

    n=len(tau)
    
    suma_sobre_clase=[]
    A=conjuntos(sigma,tau)
    for i in range(n):
        C = [[x, y] for x in A[i][0] for y in A[i][1]]
        sumanorma= sum([sgn(c[1]-c[0])*vector_principal(c[0],c[1],len(tau)) for c in C])
        suma_sobre_clase.append(sumanorma)
    return suma_sobre_clase

def suma_inercias(sigma,tau):

    n=len(tau)
    
    suma_sobre_clase=0
    A=e_conjuntos(sigma,tau)
    for i in range(n):
        suma_sobre_clase+=(n-i)*(n+1-i)*norma1(n,A[i])
    
    return suma_sobre_clase.expand()

def U2(n):
    A=representantes_de_clase(n)
    determinante=det(gram(n))

    suma=0
    for i in A:
        suma+=volumen_prueba(i[0],i[1])
    return n*(n+1)/(factorial(n+2)) *3*(suma).expand()











def M(n):

    M = [[0]*(n)]*(n)
    for i in range(n):
        M[i] = [-var('p%d%d' % tuple(sorted((i+1,j+1)))) for j in range(n)]
        M[i][i] = sum([var('p%d%d' % tuple(sorted((i+1,k))))
                       for k in range(0,n+1)]) - var('p%d%d' % (i+1,i+1))
        
    return(Matrix(M))

def d(n):
   
    lis=list(combinations(list(range(n+1)), 2))
    c=0
    for j in lis:
        i=list(j)
        c+=var('D%d%d' % tuple(sorted((i[0],i[1]))))*var('p%d%d' % tuple(sorted((i[0],i[1]))))
    return c/n







def G_conj(n):
    
    
    G = 0
    
    Ck = C_k(n+1)
    
    for g in Ck:
        k = Graph(g,multiedges=True).girth()
        if k == 2:
            G += mono(g)
        if k > 2:
            G += 2*(k-1)*mono(g)
    return(expand(G))


def mono(g):
    

    mono = 1
    for e in g:
        mono = mono*var('p%d%d' % (sorted(e)[0],sorted(e)[1]))
    return(mono)

def C_k(n):
    P = prufer(n)
    
    C = []
    
    for ps in P:
        for i in range(n):
            for j in range(i+1,n):
                c = prufer_to_tree(ps)
                c.append([i,j])
                if sorted(c) not in C:
                    C.append(sorted(c))
        
    return(C)

def prufer_to_tree(ps):
   
    n = len(ps)+2
    ps = list(ps)
    N = list(range(n))
    T = []
    while len(N) > 2:
        j = min(set(N)-set(ps))
        T.append(sorted([j,ps[0]]))
        N.remove(j)
        ps.remove(ps[0])
    T.append([N[0],N[1]])
    return(T)

def prufer(n):  

    return(itertools.product(range(n), repeat=n-2))














from concurrent.futures import ProcessPoolExecutor
import multiprocessing

def calcular_contribucion(i):
    
    return (suma_inercias(i[0], i[1]) * volumen(i[0], i[1])).expand()

def U_paralelo(n):
    start_time = time.time()
    A = representantes_de_clase(n)
    print("--- %s seconds U1 ---" % (time.time() - start_time))
    with ProcessPoolExecutor(max_workers=multiprocessing.cpu_count()) as executor:
        contribuciones = list(executor.map(calcular_contribucion, A))
    print("--- %s seconds U2 ---" % (time.time() - start_time))
    suma_total = sum(contribuciones)
    print("--- %s seconds U3 ---" % (time.time() - start_time))
    return (suma_total / (factorial(n + 2)) * 3)





def U(n):
    A=representantes_de_clase(n)
    
    start_time = time.time()
    suma=0
    for i in A:
        suma+=(suma_inercias(i[0],i[1])*volumen(i[0],i[1])).expand()
    print("--- %s seconds U1 ---" % (time.time() - start_time))
    return (suma/(factorial(n+2)) *3)



def calcular_contribucion2(i,dif,gif):
    start_tim = time.time()
    F=(dif[i[0]][i[1]] * gif[i[0]][i[1]]).expand()
    print("--- %s seconds un cálculo ---" % (time.time() - start_tim))
    return F

def final_paralelo(n):
    start_time = time.time()
    start_time = time.time()
    a=U_paralelo(n)
    
    print("--- %s seconds para U ---" % (time.time() - start_time))
    
    G=((G_conj(n)*d(n)).expand()-a)
    
    print("--- %s seconds para Gconj-U ---" % (time.time() - start_time))
    
    deter=det(gram(n)).expand()

    dif=DERIVADAS1(n,deter)
    gif=DERIVADAS2(n,G)
    print("--- %s seconds para las derivadas ---" % (time.time() - start_time))
    

    # Lista de combinaciones a dividir
    lis = [list(i) for i in list(combinations(list(range(n+1)), 2))]

    with ProcessPoolExecutor(max_workers=multiprocessing.cpu_count()) as executor:
        contribuciones = list(executor.map(calcular_contribucion2, lis,[dif]*len(lis),[gif]*len(lis)))

    suma_total = sum(contribuciones)
    print("--- %s seconds U1 ---" % (time.time() - start_time))
    return suma_total 






def final(n):
    start_time = time.time()
    a=U_paralelo(n)
    
    print("--- %s seconds para U ---" % (time.time() - start_time))
    
    G=((G_conj(n)*d(n)).expand()-a)
    
    print("--- %s seconds para Gconj-U ---" % (time.time() - start_time))
    
    deter=det(gram(n)).expand()

    dif=DERIVADAS1(n,deter)
    gif=DERIVADAS2(n,G)
    print("--- %s seconds U1 ---" % (time.time() - start_time))
    
    lis=list(combinations(list(range(n+1)), 2))
    suma=0
    for j in lis:
        i=list(j)
        suma+=(dif[i[0]][i[1]]*gif[i[0]][i[1]]).expand()
        print("--- %s seconds para un cálculo ---" % (time.time() - start_time))
        
    return suma


 
```
